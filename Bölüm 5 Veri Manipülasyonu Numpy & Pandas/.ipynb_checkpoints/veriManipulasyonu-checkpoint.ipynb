### Yeniden Şekillendirme (Reshaping)

```python
import numpy as np
```

Numpy’da yeniden şekillendirme işlemi reshape metodu ile yapılır.

```python
np.arange(1,10) # 1'den 10 a kadar sayılardan array
```

```python
np.arange(1,10).reshape((3,3))
```

Array’in içerisinde 9 tane değer olduğu için reshape ile 3'e 3'lük yeniden bir array oluşturabildik.

ilk haliyle tek boyutlu bir array’di reshape ile biz bu array’i hem 3 e 3'lük bir formata getirdik hemde boyut sayısını ikiye çıkardık.

Çalışmalarda fonksiyonların ya da döngülerin üretmiş olduğu çıktılar, tek bir boyutta tek bir array formunda gerçekleşebiliyor.
Bazen tek boyutlu array’den iki boyutluya bazende tam tersi olacak şekilde yeniden şekillendirmeye ve boyutunu değiştirmeye ihtiyacımız olabilmektedir.

```python
a = np.arange(1,10)
```

```python
a
```

```python
a.ndim # Tek boyutlu array = vector
```

Tek boyutlu array i bazen bir matris formuna çevirmek gerekebiliyor. Çevirirken tek boyuttaki bilgisi de olduğu şeklinde kalsın ve matematiksel çevrimi karşılayabilsin.

b = a.reshape((1,3)) matematiksel anlamda dönüşümü mümkün kılmayacaktır.

```python
b = a.reshape((1,9))
# 1 satır
# 9 sütun
```

```python
b.ndim # İki boyutlu array = matrix
```

```python
b
```

```python
a
```

a array’i ile b array’i arasındaki tek fark b array’inde iki tane köşeli parantez olması, bu demek oluyor ki "b array’i iki boyutlu arraydir" zaten sorguladığımız zaman bunu görebiliyoruz.

b artık bir matris tir, yani boyutu 2 dir fakat aynı şekilde tek bir vektörün taşıdığı bilgiyi taşır. Ama 1 satır 9 sütundan oluşur.

Bazen tek boyuttan çok boyuta, bazen tek boyuttan çok boyuta ama tek boyut bilgilerini taşıyacak şekilde bazı dönüştürme işlemleri ihtiyacı olabiliyor. Reshape fonksiyonu ile kolayca gerçekleştirebiliyoruz.

### Array Birlestirme (Concatenation)

```python
import numpy as np
```

Tek boyutlu array’lerde birleştirme işlemleri,

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
# liste üzerinden numpy arrayi oluşturduk
```

```python
np.concatenate([x, y])
```

```python
z = np.array([7,8,9])
```

```python
np.concatenate([x, y, z])
```

3 adet numpy array i tek boyut üzerinde yan yana gelecek şekilde birleştirdik.

İki boyutlu array’lerde birleştirme işlemleri,

```python
a = np.array([[1,2,3], 
              [4,5,6]])
# reshape fonksiyonu yerine çift köşeli parantez kullandık
```

```python
np.concatenate([a,a])
```

Satır bazında alt alta olacak şekilde aynı arrayleri birleştirdik. Eğer sütun bazında birleştirmek istersek axis argümanına 1 ekseni yani sütun bazında yan yana olacak şekilde belirterek birleştirebiliriz

axis argümanı satır ve sütün bazlı olma durumunu belirtir.

- axis=0 satır bazlı birleştirme işlemi yapar yani alt alta
- axis=1 sütün bazlı birleştirme işlemi yapar yani yanyana olacak şekilde.

```python
np.concatenate([a,a], axis = 1)
```

### Array Ayırma (Splitting)

```python
import numpy as np
```

* Tek Boyutlularda Ayırma İşlemleri

```python
x = np.array([1,2,3,99,99,3,2,1])
```

### Array Alt Küme İşlemleri ( Slicing )

Slicing işlemlerini daha önce Python yazımızda veri tiplerini incelerken ele almıştık. Burada da konu ve yöntem aynı olacak. Yani tekrar özetlemek gerekirse veri yapısı içinde belirli bi yerden belirli bir yere kadar git diyebilirim.

**Tek boyutlu slice işlemleri**

```python
import numpy as np
a = np.arange(20,30)
a
```

```python
a[:3] # yine sıfırdan üçe kadar seçme işlemini gerçekleştirme
```

### Alt Küme Üzerinde İşlem Yapmak

Array’in alt kümesine eriştikten sonra bunu isimlendirip kaydettiğimizi düşünelim.Kaydetmiş olduğumuz değişkenin üzerinde değişiklik yaptığımızda array’in ilk halinde de değişiklik olmaktadır.Ama bazen seçilen o alt küme üzerinde işlemler yapılmak istenmektedir ve array’in ilk halinde değişiklik olması istenmeyebilir.

```python
import numpy as np
a = np.random.randint(10, size=(5,5))
a
```

```python
alt_a = a[0:3, 0:2]
alt_a
```

```python
alt_a[0,0] = 99999
alt_a[1,1] = 888
```

```python
alt_a
```

```python
a
```

Yukarıda çekilen alt kümenin alt_a olarak atandığını ve sonrasında alt_a’nın sıfırıncı satır sıfırıncı sütünunda yer alan elemana 99999 değeri ile değiştirildiğini ve alt_a'nın birinci satır sıfırıncı sütununda yer alan elemana 888 değeri ile değiştirilidiğini görüyoruz. 

Sonrasında a array’ini yani arrayin orjinal ilk halini çağırdığımızda a’nin 0.satır 0.sütununda yer alan değerinde 99999 ve 1.satır 1.sütununda yer alan değerinde 888 olduğunu görüyoruz.

Bu durumda alt kümeleri bağımsızlaştırmak denilen yaklaşım uygulanması gerekiyor. En azından bu işlemlerin farkında olunması gerekiyor. Hiç farkında olunmadan bazı hatalar yapmamak adına.

Bu durum bazen çok işimize yarayabilir. Çok büyük array’lerden alt küme alıp onun üzerinde çalışıp sonrasında ana array’inde etkilenmesini istediğimiz durumlarda olabilir.

Bunun için copy() isimli bir metod kullanacağız.

```python
m = np.random.randint(10, size=(5,5))
m
```

```python
alt_b = m[0:3, 0:2].copy()
alt_b
```

```python
alt_b[0,0] = 9999
alt_b
```

```python
m
```

Not: copy() metodu slicing işleminin sonuna bağlandığında alt kümeyi ana array’den bağımsızlaştırır.

### Fancy Index ile Elemanlara Erişmek

Fancy index kavramı önemli bir kavramdır. İleride ele alacağımız pandas DataFrame yapısında hemde numpy array’lerinde daha ileri düzey eleman seçme imkanları vermektedir. Mevcut bildiğimiz eleman seçme işlemlerinden biraz daha ileri seviyedir.

Fancy herhangi bir metod veya fonksiyon olmamakla birlikte sadece bir yaklaşım biçimdir. Arka taraftaki mantığı iyi anlandığında ne zaman kullanmamız gerektiğini daha kolay görebiliyor olacağız.

```python
import numpy as np
v = np.arange(0, 30, 3)
v # Tek boyutlu array yani vector
```

```python
v[1]
```

```python
v[3]
```

```python
v[5]
```

```python
# bir arada görmek istersek
[v[1], v[3], v[5]]
```

Eğer elimizdeki bir döngü ya da fonksiyon onlarca belki yüzlerce değeri ürettiyse ve üretilen bu değerlerin her birisini gidip bir arrayin içerisinden çağırmak istiyorsam bu şekilde kaydettiğimde yani bir liste formunda arrayin içerisine koyduğumda köşeli parantez ifadenin içine bunların hepsini algılar ve seçim işlemleri gerçekleştirir.

### Numpy ile İki Bilinmeyenli Denklem Çözümü

```python
a
```

```python
b
```

```python
pd.concat([df1, df2]) # deg3 değişkeni veri setlerinden birinde olmadığı için birleştirirken problem yaratır
```

```python
pd.concat([df1, df2], join = "inner") # Kesişimlerine göre birleştirdi. 
# Yani mutlak olarak kesişimleri elemanlardan oluşan değişkenlere göre bunu birleştirdi.
```

```python
pd.concat([df1, df2], join_axes = [df1.columns])
```

```python
pd.concat([df1, df2]).reindex(columns=df1.columns)
```

Birleştirme işlemini yaptım ve taşıyan bilgi taşıyan bir değişkeni tutarak birleştirme işlemine devam ettim.  Değişkenin birisine göre birleştirme işlemini yapayım. bu birleştirme işlemini de örneğin burada seçmiş olduğumuz bir değişkenin değerlerine göre gerçekleştirebiliyoruz. df1 verisinde var3 değişkeni vardır. df2 de ise deg3 bulunmaktadır.

```python
pd.concat([df1, df2], join_axes = [df2.columns])
```

```python
pd.concat([df1, df2]).reindex(columns=df2.columns)
```

```python
pd.concat([df1, df2], join_axes = [df2.columns], ignore_index=True)
```

```python
pd.concat([df1, df2.reindex(columns = df2.columns)], ignore_index = True)
```

### Gruplama İşlemleri

Toplulaştırma işlemleri ve gruplama işlemleri bir arada düşünülmesi gereken işlemlerdir. Sözlük ile bir dataframe oluşturarark verimizi elde edelim. Key ve liste değerleriyle de bu sözlük oluşturulur.

Veri setindeki kategorik değişkenlerin gruplarının yakalanması ve bu grupların özelinde bazı işlemler yapılması için gruplama yapılır. Örneğin burada bir grup bir sınıfı temsilcisinin a sınıfı, b sınıfı, c sınıfı şeklinde ve örneğin her sınıftan da ikişer tane kişinin sınav notları elimizde olsun.

### İleri Toplulaştırma İşlemleri (Aggregate, filter, transform, apply)

```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

```python
#aggregate
```

```python
df.groupby("gruplar").mean()
```

Veri setini gruplara göre böl fakat ben öyle bir işlem istiyorum ki ne describe fonksiyonunun bana verdiği istatistikleri ne de sadece mod medyan gibi istatistikleri, kendim özellikle bir istatistik belirtmek istiyorum dediğimizde bununla ilgili de fonksiyonuyla gerçekleştirebiliyoruz. Parantezleri girdikten sonra bir liste formunda beklediği fonksiyonları verdiğimizde ve bunu bize getiriyor .

```python
df.groupby("gruplar").aggregate(["min", np.median, max])
```

```python
df.groupby("gruplar").aggregate([min, np.median, max])
```

```python
df.groupby("gruplar").aggregate(["min", np.median, "max"])
```

```python
df.groupby("gruplar").aggregate(["min", "np.median", "max"])
```

'np.median' diye bir şey bulamadı dedi. Çünkü diğerlerini algılayabiliyor fakat medyanı algılayamıyor. Yani dışsal bir fonksiyon olduğunda bunu algılayamıyor. Bu sebeple pandasın içerisinde yer aldığını bildiğimiz fonksiyonları olduğu gibi ya da tırnakla yazarsanız ki ben ayırt etmek için tırnakla yazmayı tercih ediyorum. Diğer fonksiyonları da yer aldığı kütüphaneye bağlayarak bu şekilde ifade ederseniz aggregate bunları pandasın içerisinden alır ve numpy ın içerisinde yer alan meydanı içerisinden alarak bu çalışma işlemini gerçekleştirir.

```python
df.groupby("gruplar").aggregate({"degisken1": "min", "degisken2": "max"}) 
# iki değişken içinde iki farklı istatistik uygulamak için
# degisken1 için farklı ve degisken2 için farklı fonksiyoun uyguladık.
```

### Pivot Tablolar

Satır ve sütun işlemleri yaparak veri setini amaca uygun hale getirmek için kullanılan yapılardır ve bazen groupby işlemleriyle karıştırılabilmektedir. groupby ın çok boyutlu versiyonu olarak düşünülebilir.

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby("sex")["survived"].mean() # tablosuz daha ilkel bir yöntem [] olmaması için [[]] ekle
```

[[]] ekleyerek dataframe e dönüştürdük

```python
titanic.groupby("sex")[["survived"]].mean() 
# cinsiyete göre gruplama yaparak sürekli değişkenimize "survived" a göre ortalama almak istersek
```

1 -> hayatta kalma

0 - > ölü

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean")
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

Veri setini x ekseninde "sex" kategorik değişkeni ve y ekseninde "class" kategorik değişkenin sınıflarına göre bölüp kesişimlerinde bir şeyler ifade ettik. unstack() ile hiyerarşik yapısı çözümlenir.

```python
#pivot ile table
```

```python
titanic.pivot_table("survived", index = "sex", columns = "class")
```

İndex i satırlara, class ları sütunlara koyduk.

```python
titanic.age
```

```python
titanic.age.head() # sürekli değişkeni kategorik değişkene çevirmek için boyut ekle
```

```python
age = pd.cut(titanic["age"], [0, 18, 90])
age.head(10)
```

Titanic verisi içindeki "age" değişkenini böl, aralıklar 0
18,90 değerlerine göre böl

Sonra bu bölme işlemi neticesinde çıkan sınıfları boyut olarak pivot table a ekle

```python
titanic.pivot_table("survived", ["sex", age], "class")
```

```python
titanic.pivot_table("survived", ["sex", "age"], "class") # Nan hatalı değerler verir
```

```python
titanic.pivot_table("survived", [sex, age], "class") 
# cinsiyet bu veri seti içinde yer almadığından dışardan "sex" şeklinde alınır
```
